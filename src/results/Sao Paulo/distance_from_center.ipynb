```python
import pandas as pd
from geopy.distance import geodesic

# Read the CSV files
scores_df = pd.read_csv('/workspaces/Coworking/src/results/Sao Paulo/PlacesScore.csv')
places_df = pd.read_csv('/workspaces/Coworking/src/results/Sao Paulo/SaoPaulo.csv')

# Define the center point (coordinates of downtown Sao Paulo)
center_point = (-23.5505, -46.6333)

# Calculate distances from center
places_df['Distance'] = places_df.apply(
    lambda row: geodesic(center_point, (row['Latitude'], row['Longitude'])).km,
    axis=1
)

# Merge the dataframes on the name columns
merged_df = places_df.merge(
    scores_df,
    left_on='Name',
    right_on='Place Name',
    how='inner'
)

# Display the first few rows of the merged dataframe
merged_df.head()
```

```python
# Export the merged dataframe to CSV
merged_df.to_csv('/workspaces/Coworking/src/results/Sao Paulo/merged_data.csv', index=False)
```
