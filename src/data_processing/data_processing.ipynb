```python
barcelona_score = pd.read_csv('/workspaces/Coworking/src/results/Barcelona/BarcelonaPlaces.csv')

barcelona_score.columns
```

```python
madrid_score = madrid_score[['name', 'formatted_address', 'rating', 'user_ratings_total', 'geometry.location.lat', 'geometry.location.lng']]
madrid_score.head()
```

```python
madrid_score = madrid_score.rename(columns={
    'formatted_address': 'Address',
    'geometry.location.lat': 'Latitude',
    'geometry.location.lng': 'Longitude',
    'user_ratings_total': 'User Rating Count',
    'rating': 'Rating'
})

madrid_score.head()
```

```python
madrid_merged = pd.merge(price_location, madrid_score, on='name', how='inner')

madrid_merged.head()
```

```python
all_merged = pd.concat([NewYork_merged, barcelona_merged, madrid_merged, price_location_merged], ignore_index=True)
all_merged.head()
```

```python
all_merged.shape
```

```python
all_merged.drop(columns=['Name'], inplace=True)

all_merged.head()
```

### Mean rating of each country.

```python
mean_rating_by_country = all_merged.groupby('Country')['Rating'].mean()

print(mean_rating_by_country)
```

### Best Neighboorhood of each country.

```python
# Group by Country and Neightboorhood, then calculate the weighted rating
all_merged['Weighted Rating'] = all_merged['Rating'] * all_merged['User Rating Count']
grouped = all_merged.groupby(['Country', 'Neightboorhood']).agg({
    'Weighted Rating': 'sum',
    'User Rating Count': 'sum'
}).reset_index()

# Calculate the average weighted rating
grouped['Average Weighted Rating'] = grouped['Weighted Rating'] / grouped['User Rating Count']

# Find the best neighborhood for each country
best_neighborhoods = grouped.loc[grouped.groupby('Country')['Average Weighted Rating'].idxmax()]

print(best_neighborhoods[['Country', 'Neightboorhood', 'Average Weighted Rating']])
```

## Visualizing the user rating count from the coworks of each city.

```python
import folium

# Create a map centered around New York
ny_map = folium.Map(location=[40.712776, -74.005974], zoom_start=10)

# Filter the data for New York coworking spaces
ny_coworking = all_merged[all_merged['Country'] == 'USA']

# Add bubble markers for each coworking space
for idx, row in ny_coworking.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=row['User Rating Count'] / 10,  # Adjust the divisor to change bubble size
        popup=f"{row['name']}<br>Rating: {row['Rating']}<br>User Rating Count: {row['User Rating Count']}",
        color='blue',
        fill=True,
        fill_color='blue'
    ).add_to(ny_map)

# Display the map
ny_map
```

## Score of each place.

```python
import pandas as pd

barcelona_score = pd.read_csv('/workspaces/Coworking/src/results/Barcelona/PlacesScores.csv')

madrid_score = pd.read_csv('/workspaces/Coworking/src/results/Madrid/PlacesScore.csv')

NewYork_score = pd.read_csv('/workspaces/Coworking/src/results/New York/NYC_PLaces_Score.csv')

tokyo_score = pd.read_csv('/workspaces/Coworking/src/results/Tokyo/PlacesScore.csv')

merged_scores = pd.concat([barcelona_score, madrid_score, NewYork_score, tokyo_score], ignore_index=True)

merged_scores.head()
```

```python
merged_scores.to_csv('/workspaces/Coworking/src/results/AllScores.csv', index=False)
```
